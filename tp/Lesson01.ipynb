```kotlin
val myCase = "Hello world"

when (myCase) {
    "Hello world" -> println("myCase is Hello world")
    "Hello" -> println("myCase is Hello")
    else -> println("myCase is not Hello world or Hello")
}
```

```kotlin
val pets = listOf("dog", "cat", "bird")

for (pet in pets) {
    println(pet)
}

// Tupple (A, B)
for ((index, a) in pets.withIndex()) {
    println("index: $index, pet: $a")
}

// Intervalle fermée
for (i in 1..10 step 2) {
    println(i)
}

```
